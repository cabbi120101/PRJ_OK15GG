```python
def find_champ_new_tag(gameId):
    
    global this_new_tag
    global this_all_info
    global this_cluster


    # 챔피언 원핫
    name_new_tag_win = pd.DataFrame()
    name_new_tag_lose = pd.DataFrame()

    game_info = game_match[game_match['gameId'] == gameId]
    combi_who_win = []
    combi_who_lose = []
    gameid_1 = []
    gameid_2 = []
    win = 0
    # find_team = game_info[game_info['accountId']== accountId]['who_win']
    # team_combi = game_info[game_info['who_win']==find_team]
    team_combi_A = game_info[game_info['who_win']==0]
    team_combi_B = game_info[game_info['who_win']==1]




    for i in team_combi_A['championId']:
        win = 0
        name_new_tag_lose = name_new_tag_lose.append(name_new_tag[name_new_tag['key']==i])
        combi_who_lose.append(win)
        gameid_1.append(gameId)


    for i in team_combi_B['championId']:
        win = 1
        # 챔피언 원핫
        name_new_tag_win = name_new_tag_win.append(name_new_tag[name_new_tag['key']==i])
        combi_who_win.append(win)
        gameid_2.append(gameId)



    # gameid_1.append(gameId)
    # gameid_2.append(gameId)



    if 'participantId' in team_combi_A.columns:

        name_new_tag_win['combi_win'] = combi_who_win
        name_new_tag_lose['combi_win'] = combi_who_lose
        
        name_new_tag_win.loc["onehot",:] = name_new_tag_win.sum()
        name_new_tag_lose.loc["onehot",:] = name_new_tag_lose.sum()
        

        name_new_tag_win = name_new_tag_win.loc["onehot"]
        name_new_tag_lose = name_new_tag_lose.loc["onehot"]
        name_new_tag_win = pd.DataFrame(name_new_tag_win).T
        name_new_tag_lose = pd.DataFrame(name_new_tag_lose).T

        this_champ = pd.concat([name_new_tag_win, name_new_tag_lose],ignore_index=True)

        this_champ = this_champ.drop(columns=['id','key'])


        # new_tag_lose['gameid'] = gameid_1
        # new_tag_win['gameid'] = gameid_2

        
        # this_all_info = pd.concat([this_champ, this_new_tag], axis=1)



    return this_champ
```
