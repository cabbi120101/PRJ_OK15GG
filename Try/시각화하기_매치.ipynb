### plot 준비하는 함수

```python
def prepare_plot(df):
    df_1 = df.copy()
    df_1.loc['all_sum', :] = df_1.sum()
    df_2 = df_1.loc['all_sum']
    df_2 = pd.DataFrame(df_2).T
    if 'key' in df_2.columns:
        df_2 = df_2.drop(columns=['gameid','key'])
    df_2 = df_2.T.reset_index()

    return df_2
```
